# Data Analysis and Visualization

```python
cancelled_prec = df['is_canceled'].value_counts(normalize = True)*100
cancelled_prec
```

PLOTTING THE GRAPH ON THE BASIS OF TOTAL CANCELLED AND NOT CANCELLED BOOKINGS

```python
fig = figure(figsize=(5, 4))
plt.title("Reservation  Status Count")
plt.bar(['Not canceled' , 'canceled'] , df['is_canceled'].value_counts() , edgecolor='k', width=0.7)
plt.show()
```

PLOTTING ON THE BASIS OF HOTEL TYPES (CANCELLED / NOT CANCELLED )

```python
fig = figure(figsize=(8, 4))
axl = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1, 1))

plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled' ,'canceled'])
plt.show()
```

CHECKING THE PERCENTAGE OF CANCELLED AND NOT CANCELLED BOOOKINGS FOR RESORT_HOTEL

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

zero index value shows the one which is not canceled

first index value shows the one which is CANCELED

CHECKING THE PERCENTAGE OF CANCELLED AND NOT CANCELLED BOOKINGS FOR CITY__HOTELS

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

zero index value shows the one which is not canceled

first index value shows the one which is CANCELED

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

I have created a hypothesis(situation) i.e the rate of cancellation on the resort_hotel is due to the price of the rooms

```python
plt.figure(figsize = (20,8))

plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.xlabel('Reservation_status_day')
plt.ylabel('Average daily rate')

plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x='month' , hue='is_canceled' , data=df)
plt.title('Reservation status per month', fontsize = 30)
plt.xlabel('month')
plt.ylabel('Number Of Reservation')
plt.legend(['NOT CANCELED' ,'CANCELED'])

plt.show
```

PLOTTING ADR FOR EACH MONTH.

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), estimator=sum)
plt.legend(fontsize=20)
plt.show()
```

This plot shows that in january the adr was high and in august it is low.

this is the reason the cancellation rate is highjer in the month of january and it was low in the month of august.

# COUNTRY WISE CANCELLATION RATE

```python
df['country'].unique()
```

```python
cancelled_data = df[df['is_canceled'] == 1]

top_1_country = cancelled_data[ 'country' ].value_counts()


plt.figure(figsize = (8,8))

plt.title(' Countries with reservation canceled')
plt.pie(top_1_country, autopct = '%.2f', labels = top_1_country. index)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]

top_10_country = cancelled_data[ 'country' ].value_counts()[:10]


plt.figure(figsize = (8,8))

plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country. index)
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))

plt.title('Average Daily Rate')

plt.plot(not_cancelled_df_adr[ 'reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```
